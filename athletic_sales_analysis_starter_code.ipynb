```python
# Import Libraries and Dependencies
import pandas as pd
from datetime import date
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020 = pd.read_csv("Resources/athletic_sales_2020.csv") # read the files into a variables
athletic_sales_2021 = pd.read_csv("Resources/athletic_sales_2021.csv")
sales_2020 = pd.DataFrame(athletic_sales_2020) # converting the variables to a data frame
sales_2021 = pd.DataFrame(athletic_sales_2021)
```

```python
# Display the 2020 sales DataFrame
display(sales_2020)
```

```python
# Display the 2021 sales DataFrame
display(sales_2021)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
sales_2020_2021_df = pd.concat([sales_2020, sales_2021], axis='rows', join='inner')
sales_2020_2021_df = sales_2020_2021_df.reset_index(drop=True)
display(sales_2020_2021_df)
```

```python
# Check if any values are null.
sales_2020_2021_df.isnull().sum()
```

```python
# Check the data type of each column
sales_2020_2021_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
# use format optio

# .to_datetime('2018-10-26 12:00:00.0000000011',
# ...                format='%Y-%m-%d %H:%M:%S.%f')
# sales_2020_2021_df['invoice_date'] = pd.to_datetime(sales_2020_2021_df['invoice_date'])
# sales_2020_2021_df['invoice_date'].to_datetime(['invoice_date'], format='%m/%d/%y', errors="coerce")
sales_2020_2021_df['invoice_date'] = pd.to_datetime(sales_2020_2021_df['invoice_date'], errors="coerce")
sales_2020_2021_df.head(20)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
total_sold_products = sales_2020_2021_df.groupby(['region', 'state', 'city'])[['units_sold']].sum()
# Rename the sum to "Total_Products_Sold".
total_sold_products = total_sold_products.rename(columns={"units_sold":"Total_Products_Sold"})

# Show the top 5 results.
total_sold_products.sort_values(by=['Total_Products_Sold'], ascending=False).head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
Total_Products_Sold = pd.pivot_table(sales_2020_2021_df, index=['region', 'state', 'city'], values=['units_sold'], aggfunc='sum', sort='descending')

# Rename the "units_sold" column to "Total_Products_Sold"
Total_Products_Sold = Total_Products_Sold.rename(columns={"units_sold":"Total_Products_Sold"})

# Show the top 5 results.
Total_Products_Sold.sort_values(by=['Total_Products_Sold'], ascending=False).head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
womans_top_day_df = pd.pivot_table(womans_athletic_footware_df, index=['invoice_date'], values=['total_sales'], aggfunc='sum', sort='descending')
# Optional: Rename the "total_sales" column to "Total Sales"
womans_top_day_df_rn = womans_top_day_df.rename(columns={"total_sales":"Total Sales"})

# Show the table.
womans_top_day_df_rn.sort_values(by=["Total Sales"], ascending=False).head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# create and empty dictionary for each day
monday_total = 0
tuesday_total = 0
wednesday_total = 0
thursday_total = 0
friday_total = 0
saturday_total = 0
sunday_total = 0
# cycle through each roe of the dataframe and check which day of the week it is
for womans_top_day_df['invoice_date'] in womans_top_day_df:
    if womans_top_day_df['invoice_date'].weekday() == 0:
        monday_total += womans_top_day_df['Total Sales']
    elif womans_top_day_df['invoice_date'].weekday() == 1:
        tuesday_total += womans_top_day_df['Total Sales']
    elif womans_top_day_df['invoice_date'].weekday() == 2:
        wednesday_total += womans_top_day_df['Total Sales']
    elif womans_top_day_df['invoice_date'].weekday() == 3:
        thursday_total += womans_top_day_df['Total Sales']
    elif womans_top_day_df['invoice_date'].weekday() == 4:
        friday_total += womans_top_day_df['Total Sales']
    elif womans_top_day_df['invoice_date'].weekday() == 5:
        saturday_total += womans_top_day_df['Total Sales']
    else:
        sunday_total += womans_top_day_df['Total Sales']
#make a new dataframe with the totals from the previous section

womans_total_df = pd.DataFrame({'Monday': monday_total, 'Tuesday': tuesday_total, 'Wednesday': wednesday_total, 'Thursday': thursday_total, 'Friday': friday_total, 'Saturday': saturday_total, 'Sunday': sunday_total})

# Sort the resampled pivot table in ascending order on "Total Sales".
womans_total_df.sort_values(by=["Total Sales"], ascending=False).head()
```
